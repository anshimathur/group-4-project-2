Beginning the import of the csv and converting to a dataframe.

```python
import pandas as pd

raw_data = pd.read_csv("resources/Life Expectancy Data.csv")
# Display the first few rows
raw_data.head()
```

```python
#Taking a high level initial look at the data frame.
raw_data.info()
```

```python
#Some columns have extra spaces (like " thinness 1-19 years").  Cleaning these by removing extra space.
raw_data.columns = raw_data.columns.str.strip()
# Display the first few rows (hard to discern the difference in the output, but the extra spaces have been removed)
raw_data.head()
```

```python
#Some columns have trailing spaces in the values.  Cleaning these by removing extra space.
raw_data.columns = raw_data.columns.str.strip()
raw_data.head()
```

```python
#Determining the percentage of missing values in each column.
missing_percent = raw_data.isnull().mean() * 100
missing_percent
```

```python
#No column has >30% missing values so no need to drop any columns.  There are a small number of rows that do not contain 'life expectancy' values.
#Code to drop rows that do not contain 'life expectancy' values.
#Renumbering index after dropping rows.
raw_data = raw_data.dropna(subset=['Life expectancy']).reset_index(drop=True)
raw_data.info()
```

```python
#Impute missing feature values for columns with null values remaining.
#Whether to use mean or median depends on the distribution of the data.  If the data is normally distributed, use mean.  If the data is skewed, use median.
# Step 1: Select numerical columns
num_cols = raw_data.select_dtypes(include='number').columns

# Step 2: Find which of those columns have missing values
num_cols_with_na = [col for col in num_cols if raw_data[col].isnull().sum() > 0]

# Step 3: Generate histograms with columns that have missing values
import matplotlib.pyplot as plt

# Step 3: Plot distributions
for col in num_cols_with_na:
    raw_data[col].hist(bins=50)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
#Only'Total expenditure' and 'Schooling' have a normal distribution.  We will use mean to impute missing values for those two features
#For the other columns, we will use the median to impute missing values.
mean_impute_cols = ['Total expenditure', 'Schooling']

# Step 3: Apply imputation
for col in num_cols_with_na:
    if col in mean_impute_cols:
        raw_data[col] = raw_data[col].fillna(raw_data[col].mean())
    else:
        raw_data[col] = raw_data[col].fillna(raw_data[col].median())
#Verify that there are no more missing values
raw_data.info()
```

```python
#Encoding the categorical features.  First, how many unique values are in each categorical feature?
# Select categorical columns
cat_cols = raw_data.select_dtypes(include='object').columns

# Count unique values in each
for col in cat_cols:
    unique_vals = raw_data[col].nunique()
    print(f"{col}: {unique_vals} unique values")
```

```python
#Drop county as it has too many unique values to be encoded usefully.
raw_data = raw_data.drop(columns=['Country'])
raw_data.info()
```

```python
#Encode "Status" using one-hot encoding.
# Use get_dummies and extract the one dummy column (because it's binary)
status_encoded = pd.get_dummies(raw_data['Status'], drop_first=True)

# Replace the original 'Status' column with the encoded values
raw_data['Status'] = status_encoded
```

```python
#Use StandardScaler for numeric columns (excluding the target):

from sklearn.preprocessing import StandardScaler

# Separate target
target = raw_data['Life expectancy']

# Select only numeric features (excluding the target)
numeric_cols = raw_data.select_dtypes(include='number').columns.drop('Life expectancy')

# Initialize scaler and fit-transform the numeric columns
scaler = StandardScaler()
scaled_features = scaler.fit_transform(raw_data[numeric_cols])

# Replace the numeric columns in raw_data with their scaled versions
raw_data[numeric_cols] = scaled_features
raw_data
```
